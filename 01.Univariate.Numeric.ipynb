# Univariate: Numeric 

```r
require(tidyverse)
require(scales)
```

```r
nhanes <- read_csv("Datasets/nhanes.csv")
```

```r
#nhanes %>% head()
```

# Age

### Distribution - Histogram

```r
nhanes %>% 
    ggplot() + 
        geom_histogram(aes(x=Age), fill='orange', col='white', bins=20) + 
        scale_y_continuous(n.breaks=10) + 
        labs(title='Histogram showing age distribution in NHANES dataset') + 
        theme(title=element_text(size=13.5, face='bold'))
```

### Distribution - Density plot

```r
nhanes %>% 
    ggplot() + 
        geom_density(aes(x=Age), fill='orange', col='white') + 
        labs(title='Density plot showing age distribution in NHANES dataset') + 
        theme(title=element_text(size=13.5, face='bold'))
```

# Non normal data

### Numeric Summaries

```r
nhanes %>% 
    with(Age) %>% 
    fivenum()

#Min=0, Q1=10, Median=28, Q3=53, Max=80
```

```r
nhanes %>% 
    with(Age) %>% 
    IQR()
#IQR=43
```

### Boxplot - horizontal

```r
nhanes %>% 
    ggplot() + 
        geom_boxplot(aes(x=Age), fill='orange') +        
        labs(title='Horizontal boxplot of age in NHANES dataset') + 
        theme(title=element_text(size=13.5, face='bold'), axis.text.y = element_blank(), axis.ticks.y = element_blank(), axis.text.x = element_text(size=16))
```

### Boxplot - vertical

```r
nhanes %>% 
    ggplot() + 
    geom_boxplot(aes(y=Age), fill='orange') +
    labs(title='Vertical boxplot of age in NHANES dataset') + 
    theme(
        title=element_text(size=13.5, face='bold'), 
        axis.text.x = element_blank(), 
        axis.ticks.x=element_blank(), 
        axis.text.y=element_text(size=16), 
        axis.title.y = element_text(size=16)
    )
```

# Systolic Blood Pressure (contains some missings)

### Distribution - histogram

```r
nhanes %>% 
    filter(!is.na(BPSysAve)) %>%
    ggplot() + 
        geom_histogram(aes(x=BPSysAve), col='white', fill='orange', bins=20) + 
        labs(title='Histogram of Systolic Blood Pressure', x='Systolic Blood Pressure')
```

### Distribution: Density plot

```r
nhanes %>% 
    filter(!is.na(BPSysAve)) %>%
    ggplot() + 
        geom_density(aes(x=BPSysAve), fill='orange', col='white') + 
        labs(title='Density plot of Systolic Blood Pressure', x='Systolic Blood Pressure')
```

# Normal Data

### Numeric Summaries for Center and Spread

```r
nhanes %>% 
    with(BPSysAve) %>% 
    mean(na.rm=T) %>% 
    round(2)
```

```r
nhanes %>% 
    with(BPSysAve) %>% 
    sd(na.rm=T) %>% 
    round(2)
```

```r
nhanes %>% 
    filter(!is.na(BPSysAve)) %>%
    ggplot() + 
        geom_boxplot(aes(x=BPSysAve), fill='orange') + 
        labs(title='Distribution of Systolic Blood Pressure', x='Systolic Blood Pressure')
```

### Boxplot - vertical

```r
nhanes %>% 
    filter(!is.na(BPSysAve)) %>%
    ggplot() + 
        geom_boxplot(aes(y=BPSysAve), fill='orange') + 
        labs(title='Distribution of Systolic Blood Pressure', y='Systolic Blood Pressure')
```

# Combining Histogram with Density plot

```r
nhanes %>% 
    filter(!is.na(BPSysAve)) %>%
    ggplot(aes(x=BPSysAve)) + 
        geom_histogram(aes(y=after_stat(density)), col='white', fill='orange', bins=20, position='identity') + 
        geom_density(fill='blue', col='navyblue', alpha=.2) 
```
