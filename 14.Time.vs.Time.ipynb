# Bivariate: Time vs Time

```r
require(tidyverse)
require(repr)
```

```r
infection <- read_csv("Datasets/infection.csv")
```

# Compute intervals

```r
infection %>% 
    mutate(duration = end_date - start_date) %>% 
    head()
```

# Use Gantt plots to represent the duration

```r
options(repr.plot.width=16, repr.plot.height=12)
```

```r
infection %>% 
    filter(start_date >='2004-04-30' & end_date <= '2004-06-30') %>%
    ggplot() + 
        geom_segment(aes(x=start_date, xend=end_date, y=person, yend=person, col=person)) + 
        geom_point(aes(x=start_date, y=person, col=person), size=5) + 
        geom_point(aes(x=end_date, y=person, col=person), size=5)
        
```
