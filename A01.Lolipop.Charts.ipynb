# Appendix 1: Lollipop Plots

```r
require(tidyverse)
```

```r
nhanes <- read_csv('Datasets/nhanes.csv')
```

```r
nhanes %>% 
    group_by(Race1) %>% 
    summarise(average_height = mean(Height, na.rm=T), tallest=round(max(Height, na.rm=T))) %>% 
    ggplot() + 
        geom_segment(aes(x=Race1, xend=Race1, y=0, yend=tallest), col='navyblue') +         
        geom_point(aes(x=Race1, y=tallest), size=5, col='navyblue') 
        
```

```r
nhanes %>% 
    group_by(Race1) %>% 
    summarise(average_height = mean(Height, na.rm=T), tallest=round(max(Height, na.rm=T))) %>% 
    ggplot() + 
        geom_segment(aes(y=Race1, yend=Race1, x=0, xend=tallest), col='navyblue') +         
        geom_point(aes(y=Race1, x=tallest), size=5, col='navyblue') 
        
```

```r
nhanes %>% 
    group_by(Race1) %>% 
    summarise(average_height = mean(Height, na.rm=T), tallest=round(max(Height, na.rm=T))) %>% 
    ggplot() + 
        geom_segment(aes(x=Race1, xend=Race1, y=0, yend=tallest)) +         
        geom_col(aes(x=Race1,  y=average_height)) + 
        geom_point(aes(x=Race1, y=tallest), size=5)
```

```r
nhanes %>% 
    group_by(Race1) %>% 
    summarise(average_height = mean(Height, na.rm=T), tallest=round(max(Height, na.rm=T))) %>% 
    ggplot() + 
        geom_segment(aes(x=Race1, xend=Race1, y=0, yend=tallest, col='orange')) +         
        geom_col(aes(x=Race1,  y=average_height), fill='orange') + 
        geom_point(aes(x=Race1, y=tallest), size=10, alpha=0.5, col='orange') +
        geom_point(aes(x=Race1, y=tallest), size=8, col='orange') + 
        theme_bw()
        
```
