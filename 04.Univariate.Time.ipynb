# Univariate: Time

```r
require(tidyverse)
```

```r
health <- read_csv("Datasets/mental_health.csv")
```

# Time

```r
health %>% 
    arrange(Timestamp) %>%
    group_by(year(ymd_hms(Timestamp))) %>% 
    summarise(Frequency = n()) %>%
    mutate(Year = `year(ymd_hms(Timestamp))`) %>%
    ggplot() + geom_col(aes(x=Year, y=Frequency), fill='orange', col='white')
```

### Alteratives 
You may use other categorical plots for time as a univariate.
