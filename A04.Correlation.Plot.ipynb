# Appendix 4: Correlation Heatmap

```r
require(tidyverse)
require(reshape2)
```

```r
nhanes <- read_csv("Datasets/nhanes.csv")
```

# Correlation Plots - Heatmaps

```r
nhanes %>% 
    select(Age, Height, Weight, BMI, Length) %>% 
    cor(use = 'complete.obs') %>% 
    melt() %>% 
    select(X=Var1, Y=Var2, value) %>% 
    ggplot(aes(x=X, y=Y, fill=value, label=round(value, 2))) + 
        geom_tile() + 
        scale_fill_gradient(low='white', high='red') + 
        geom_text(col='black')
```
