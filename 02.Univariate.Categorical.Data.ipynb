```r
require(tidyverse)
```

```r
nhanes <- read_csv("Datasets/nhanes.csv")
INEC <- read_csv("Datasets/INEC.csv")
```

# Race

### Frequency Plots - barcharts

```r
nhanes %>% 
    group_by(Race1) %>% 
    summarise(Frequency = n()) %>% 
    mutate(Proportion = round(Frequency/sum(Frequency), 2)) %>% 
    mutate(Percentage = round(Frequency*100/sum(Frequency), 2))
```

### You can plot for Frequency, Proportion or Prcentage

```r
nhanes %>% 
    ggplot() + 
        geom_bar(aes(x=Race1), fill='orange', col='white')
```

```r
#especially when there are too many levels
INEC %>% 
    ggplot() + 
        geom_col(aes(y=State, x=APC), fill='orange', col='white')
```

```r
nhanes %>% 
    ggplot() + 
        geom_bar(aes(x=Race1, fill=Race1), color='white') + 
        coord_polar('x') +
        labs(x='', y='') + 
        theme(axis.text = element_blank(), axis.ticks = element_blank())
```

```r
nhanes %>% 
    ggplot() + 
        geom_bar(aes(x=Race1, fill=Race1), width=1, color='white') + 
        coord_polar('x') +
        labs(x='', y='') + 
        theme(axis.text = element_blank(), axis.ticks = element_blank())
```

```r
nhanes %>% 
    ggplot() + 
        geom_bar(aes(x=Race1, fill=Race1), width=1, color='white') + 
        coord_polar('y') + 
        theme(axis.text = element_blank(), axis.ticks = element_blank()) + 
        labs(x='', y='')
```

```r
nhanes %>% 
    group_by(Race1) %>% 
    summarise(Frequency=n()) %>% 
    ggplot() +
        geom_col(aes(x=4, y=Frequency, fill=Race1), col='white', lwd=2) + 
        coord_polar(theta='y') + 
        xlim(c(2.5, 4.5))
        
```

```r
nhanes %>% 
    group_by(Race1) %>% 
    summarise(Frequency=n()) %>% 
    ggplot() +
        geom_col(aes(x=Frequency, y='', fill=Race1), col='white') + 
        coord_polar(start=0) + 
        theme(axis.text = element_blank(), axis.ticks = element_blank()) + 
        labs(x='', y='')
```

```r
nhanes %>% 
    group_by(Race1) %>% 
    summarise(Frequency=n()) %>% 
    ggplot(aes(x=Frequency, y='', label=Frequency, fill=Race1)) + 
        geom_col(col='white') + 
        coord_polar('x', start=0) + 
        geom_text(position=position_stack(vjust=0.5), col='white') + 
        theme(axis.text = element_blank(), axis.ticks = element_blank()) + 
        labs(x='', y='')
```
