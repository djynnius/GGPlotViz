# Bivariate: Geographic vs Time

```r
require(tidyverse)
require(repr)
require(scales)
```

```r
gdp <- read_csv("Datasets/GDP.csv")
```

# Trend line

```r
options(repr.plot.width=20, repr.plot.height=8)
```

```r
gdp %>% 
    filter(`Country Name` %in% c('Nigeria', 'Ghana', 'South Africa', 'Egypt', 'Kenya', 'Belgium')) %>% 
    pivot_longer(cols=as.character(1960:2020), names_to='Year', values_to='GDP') %>% 
    mutate(Year = as.numeric(Year)) %>% 
    ggplot() + 
        geom_line(aes(y=GDP, x=Year, group=`Country Name`, col=`Country Name`), lwd=0.75) + 
        geom_point(aes(y=GDP, x=Year, col=`Country Name`)) + 
        scale_y_continuous(labels = label_number(suffix='b', scale = 1e-9, prefix = '$'), n.breaks=8) + 
        scale_x_continuous(n.breaks=10)
```

# Area Plot

```r
gdp %>% 
    filter(`Country Name` %in% c('Nigeria', 'Ghana', 'South Africa', 'Egypt', 'Kenya', 'Belgium')) %>% 
    pivot_longer(cols=as.character(1960:2020), names_to='Year', values_to='GDP') %>% 
    mutate(Year = as.numeric(Year)) %>% 
    ggplot() + 
        geom_area(aes(y=GDP, x=Year, group=`Country Name`, fill=`Country Name`), col='white', lwd=0.25) + 
        scale_y_continuous(labels = label_number(suffix='b', scale = 1e-9, prefix = '$'), n.breaks=8) + 
        scale_x_continuous(n.breaks=10)
```
