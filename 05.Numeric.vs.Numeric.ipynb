# Bivariate: Numeric vs Numeric

```r
require(tidyverse)
```

```r
nhanes <- read_csv("Datasets/nhanes.csv")
```

```r
nhanes %>% 
    filter(!is.na(Weight) & !is.na(Height)) %>% 
    filter(Age %in% 50:60) %>% 
    ggplot() + 
        geom_point(aes(x=Weight, y=Height))
```

```r
nhanes %>% 
    filter(!is.na(Weight) & !is.na(Height)) %>% 
    filter(Age %in% 50:60) %>% 
    ggplot() + 
        geom_smooth(aes(x=Weight, y=Height), method='lm')
```

```r
nhanes %>% 
    filter(!is.na(Weight) & !is.na(Height)) %>% 
    filter(Age %in% 50:60) %>% 
    ggplot() + 
        geom_point(aes(x=Weight, y=Height)) +
        geom_smooth(aes(x=Weight, y=Height), method='lm')
```
